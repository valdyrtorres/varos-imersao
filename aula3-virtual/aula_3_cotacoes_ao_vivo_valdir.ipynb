```python
tempo = time.time() + 10

while time.time() < tempo:

    tick = mt5.symbol_info_tick("PETR4")
    print(f"O fechamento é {tick.last}")
    print(f"O valor de compra é {tick.ask}")
    print(f"O valoe de venda é {tick.bid}")
    print("-----------------------------")
    time.sleep(1)
    
```
